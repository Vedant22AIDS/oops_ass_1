```python
'''
Que12
Abstract Base Classes (ABCs) in Python are special types of classes that define a set of abstract 
methods and attributes that its subclasses must implement. ABCs are used to provide a blueprint for
concrete classes, and they cannot be instantiated directly.ABCs are defined using the abc module. 
To define an ABC, you use the ABC metaclass, and you can decorate methods as abstract using the
@abstractmethod decorator.
'''
from abc import ABC,abstractmethod
class operation(ABC):
    @abstractmethod
    def add(self):
        pass
    @abstractmethod
    def sub(self):
        pass
    @abstractmethod
    def mul(self):
        pass
    @abstractmethod
    def div(self):
        pass

class Arithmetic(operation):
    def add(self,a,b):
        return a+b
    def sub(self,a,b):
        return a-b
    def mul(self,a,b):
        return a*b
    def div(self,a,b):
        return a//b
op=Arithmetic()
print(op.add(10,20))
print(op.sub(10,20))
print(op.mul(10,20))
print(op.div(10,20))
```

```python
#Encapsulation
'''
Que1
Encapsulation is a fundamental concept of object-oriented programming (OOP).
It involves bundling data and methods into a single unit, called a class. This bundling process 
helps to restrict direct access to the data and methods, and promotes better code organization and
reusability.

Que2
Encapsulation is one of the fundamental principles of object-oriented programming (OOP) that focuses on bundling the data (attributes or properties) and the methods (functions or procedures) that operate on the data into a single unit called a class. This concept is crucial for creating modular, reusable, and maintainable code. Two key aspects of encapsulation are access control and data hiding.

1. **Access Control:**
   - **Public:** Members (attributes or methods) declared as public are accessible from outside the class. They form the interface through which external code interacts with the class. Public members are essential for allowing other classes or objects to use the functionality provided by the class.
  
   - **Private:** Members declared as private are only accessible within the class itself. They are hidden from external access, providing a level of security and preventing direct manipulation of the internal state by external entities. Private members are typically accessed and modified through public methods, allowing the class to control and validate the interactions with its internal data.

   - **Protected:** Members declared as protected are accessible within the class and its subclasses (derived classes). This allows for a controlled level of access, enabling the subclass to inherit and use certain members while still restricting direct access from external code.

2. **Data Hiding:**
   - Data hiding involves restricting the access to the internal details of an object. The internal state (attributes or properties) of an object should not be directly accessible from outside the class. Instead, external code should interact with the object through well-defined interfaces provided by public methods.
  
   - By hiding the implementation details, a class can maintain a stable external interface while making changes to its internal structure without affecting the external code that uses it. This concept promotes code flexibility, as modifications to the internal implementation do not necessitate changes to the code outside the class.

In summary, encapsulation, through access control and data hiding, helps in creating modular and secure code by managing the visibility and accessibility of class members. It promotes the principle of information hiding, where the internal details of a class are concealed, and interactions are conducted through a well-defined and controlled interface. This enhances code organization, reusability, and maintenance in the context of object-oriented programming.
'''
```

```python
#Que11
class Person:
    def __init__(self, name, age, gender):
        self.__name = name
        self.__age = age
        self.__gender = gender

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def get_gender(self):
        return self.__gender

    def set_name(self, new_name):
        self.__name = new_name

    def set_age(self, new_age):
        self.__age = new_age

    def set_gender(self, new_gender):
        self.__gender = new_gender


class Student(Person):
    def __init__(self, name, age, gender, student_id):
        super().__init__(name, age, gender)
        self.__student_id = student_id
        self.__courses_enrolled = []

    def get_student_id(self):
        return self.__student_id

    def get_courses_enrolled(self):
        return self.__courses_enrolled

    def enroll_in_course(self, course):
        self.__courses_enrolled.append(course)

    def drop_course(self, course):
        if course in self.__courses_enrolled:
            self.__courses_enrolled.remove(course)


class Teacher(Person):
    def __init__(self, name, age, gender, employee_id):
        super().__init__(name, age, gender)
        self.__employee_id = employee_id
        self.__courses_taught = []

    def get_employee_id(self):
        return self.__employee_id

    def get_courses_taught(self):
        return self.__courses_taught

    def add_course_taught(self, course):
        self.__courses_taught.append(course)

    def remove_course_taught(self, course):
        if course in self.__courses_taught:
            self.__courses_taught.remove(course)


class Course:
    def __init__(self, course_code, course_name):
        self.__course_code = course_code
        self.__course_name = course_name
        self.__students_enrolled = []
        self.__teacher_assigned = None

    def get_course_code(self):
        return self.__course_code

    def get_course_name(self):
        return self.__course_name

    def get_students_enrolled(self):
        return self.__students_enrolled

    def get_teacher_assigned(self):
        return self.__teacher_assigned

    def set_teacher_assigned(self, teacher):
        self.__teacher_assigned = teacher

    def enroll_student(self, student):
        self.__students_enrolled.append(student)

    def remove_student(self, student):
        if student in self.__students_enrolled:
            self.__students_enrolled.remove(student)


# Example Usage:

# Create students, teachers, and courses
student1 = Student(name="Alice", age=18, gender="Female", student_id="S001")
student2 = Student(name="Bob", age=17, gender="Male", student_id="S002")

teacher1 = Teacher(name="Ms. Johnson", age=35, gender="Female", employee_id="T001")
teacher2 = Teacher(name="Mr. Smith", age=40, gender="Male", employee_id="T002")

math_course = Course(course_code="MATH101", course_name="Mathematics")
english_course = Course(course_code="ENG201", course_name="English")

# Enroll students and assign teachers to courses
math_course.enroll_student(student1)
math_course.enroll_student(student2)
math_course.set_teacher_assigned(teacher1)

english_course.enroll_student(student1)
english_course.enroll_student(student2)
english_course.set_teacher_assigned(teacher2)

# Display information
print(f"{student1.get_name()} (Student ID: {student1.get_student_id()}) is enrolled in {math_course.get_course_name()} taught by {teacher1.get_name()}")
print(f"{student2.get_name()} (Student ID: {student2.get_student_id()}) is enrolled in {english_course.get_course_name()} taught by {teacher2.get_name()}")
```

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Withdrawal of ${amount} successful. Remaining balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal failed.")

    def display_info(self):
        print(f"Account Number: {self.account_number}")
        print(f"Balance: ${self.balance}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, interest_rate):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def withdraw(self, amount):
        # Special handling for SavingsAccount withdrawals, considering interest rate
        if amount <= self.balance:
            self.balance -= amount
            self.balance -= self.balance * (self.interest_rate / 100)
            print(f"Withdrawal of ${amount} from Savings Account successful. Remaining balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal failed.")


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance, overdraft_limit):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        # Special handling for CheckingAccount withdrawals, considering overdraft limit
        total_funds = self.balance + self.overdraft_limit
        if amount <= total_funds:
            self.balance -= amount
            print(f"Withdrawal of ${amount} from Checking Account successful. Remaining balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal failed.")


class CreditCardAccount(BankAccount):
    def __init__(self, account_number, balance, credit_limit):
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        # Special handling for CreditCardAccount withdrawals, considering credit limit
        total_credit = self.balance + self.credit_limit
        if amount <= total_credit:
            self.balance -= amount
            print(f"Withdrawal of ${amount} from Credit Card successful. Remaining balance: ${self.balance}")
        else:
            print("Insufficient funds. Withdrawal failed.")


# Example usage:
savings_account = SavingsAccount(account_number="SA123", balance=1000, interest_rate=2.5)
checking_account = CheckingAccount(account_number="CA456", balance=1500, overdraft_limit=500)
credit_card_account = CreditCardAccount(account_number="CC789", balance=-200, credit_limit=1000)

accounts = [savings_account, checking_account, credit_card_account]

for account in accounts:
    account.display_info()
    account.withdraw(200)
    print()
```

```python
'''
Que5
Abstract classes in Python are classes that cannot be instantiated directly. Instead, they're intended to be subclassed by other classes that provide concrete implementations of its abstract methods. Abstract classes are useful for defining a common interface for a group of related classes.
Regular classes, on the other hand, can be instantiated directly and can have both abstract and concrete methods. Concrete methods are methods that have a complete implementation.

Que7
In object-oriented programming (OOP), an interface is a reference type that defines a set of abstract methods. An abstract method is a method that has no implementation. Classes that implement an interface must provide concrete implementations for all of the interface's abstract methods.
Interfaces are used to achieve abstraction, which is the process of hiding the implementation details of a class from its users. This allows users to focus on what a class does, rather than how it does it.
To create an interface in Python, you use the abc module. The abc module provides a number of classes and functions that can be used to create and implement interfaces.
'''
#Que6
class BankAccount:
    def __init__(self, account_number, account_holder):
        self._account_number = account_number  # Private attribute
        self._account_holder = account_holder  # Private attribute
        self._balance = 0  # Private attribute, initialized to 0

    def deposit(self, amount):
        """Deposit funds into the account."""
        if amount > 0:
            self._balance += amount
            print(f"Deposited ${amount}. New balance: ${self._balance}")
        else:
            print("Invalid deposit amount. Please deposit a positive amount.")

    def withdraw(self, amount):
        """Withdraw funds from the account."""
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self._balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        """Get the current balance."""
        return self._balance

    def display_info(self):
        """Display account information."""
        print(f"Account Number: {self._account_number}")
        print(f"Account Holder: {self._account_holder}")
        print(f"Balance: ${self._balance}")

# Example usage:
account = BankAccount(account_number="123456", account_holder="John Doe")

account.display_info()  # Display initial account information

account.deposit(1000)  # Deposit $1000
account.withdraw(500)  # Withdraw $500

print(f"Current Balance: ${account.get_balance()}")
```

```python
#Que8
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name, species):
        self.name = name
        self.species = species

    @abstractmethod
    def make_sound(self):
        pass

    @abstractmethod
    def eat(self):
        pass

    @abstractmethod
    def sleep(self):
        pass

class Mammal(Animal):
    def __init__(self, name, species, fur_color):
        super().__init__(name, species)
        self.fur_color = fur_color

    def make_sound(self):
        return "Mammals make various sounds."

class Bird(Animal):
    def __init__(self, name, species, feather_color):
        super().__init__(name, species)
        self.feather_color = feather_color

    def make_sound(self):
        return "Birds chirp and sing."

class Dog(Mammal):
    def __init__(self, name, fur_color, breed):
        super().__init__(name, species="Dog", fur_color=fur_color)
        self.breed = breed

    def eat(self):
        return f"{self.name} the {self.breed} is eating dog food."

    def sleep(self):
        return f"{self.name} the {self.breed} is sleeping in a cozy bed."

class Parrot(Bird):
    def __init__(self, name, feather_color, breed):
        super().__init__(name, species="Parrot", feather_color=feather_color)
        self.breed = breed

    def eat(self):
        return f"{self.name} the {self.breed} is eating seeds."

    def sleep(self):
        return f"{self.name} the {self.breed} is resting on a perch."

# Example usage:
dog = Dog(name="Buddy", fur_color="Brown", breed="Golden Retriever")
parrot = Parrot(name="Polly", feather_color="Green", breed="African Grey")

animals = [dog, parrot]

for animal in animals:
    print(f"{animal.name} ({animal.species}) says: {animal.make_sound()}")
    print(animal.eat())
    print(animal.sleep())
    print()
```

```python
'''
Que9
Significance of Encapsulation in Achieving Abstraction:
Data Hiding:

Encapsulation allows the implementation details of an object to be hidden from the outside world.
The internal state of an object is encapsulated within the class, and access to it is controlled through methods.
This helps in preventing direct manipulation of object properties and ensures that the object's state remains consistent.
Abstraction:

Abstraction is the process of simplifying complex systems by modeling classes based on the essential properties and behaviors they possess.
Encapsulation facilitates abstraction by exposing only relevant details through well-defined interfaces (public methods), while hiding unnecessary complexities.
Users of the class interact with the object through a set of methods, abstracting away the internal implementation.
Modularity:

Encapsulation promotes modularity, as each class acts as a self-contained unit with its own internal implementation.
Changes to the internal implementation of a class do not affect external code as long as the public interface remains unchanged.
This modularity makes it easier to maintain and update code without impacting the entire system.
'''
class Person:
    def __init__(self, name, age):
        self._name = name  # Private attribute
        self._age = age    # Private attribute

    def get_name(self):
        return self._name

    def set_name(self, new_name):
        self._name = new_name

    def get_age(self):
        return self._age

    def set_age(self, new_age):
        if 0 < new_age < 120:  # Validation for age
            self._age = new_age
        else:
            print("Invalid age!")

# Example usage:
person = Person(name="Alice", age=25)

print(person.get_name())  # Accessing name through a getter method
person.set_name("Bob")    # Modifying name through a setter method
print(person.get_name())

print(person.get_age())   # Accessing age through a getter method
person.set_age(30)        # Modifying age through a setter method
print(person.get_age())
```

```python
#Que13
from abc import ABC, abstractmethod

class TeamMember(ABC):
    def __init__(self, name, role):
        self.name = name
        self.role = role

    @abstractmethod
    def get_details(self):
        pass

class Manager(TeamMember):
    def __init__(self, name, team_size):
        super().__init__(name, "Manager")
        self.team_size = team_size

    def get_details(self):
        return f"{self.role}: {self.name}, Team Size: {self.team_size}"

class Developer(TeamMember):
    def __init__(self, name, programming_language):
        super().__init__(name, "Developer")
        self.programming_language = programming_language

    def get_details(self):
        return f"{self.role}: {self.name}, Programming Language: {self.programming_language}"

class Designer(TeamMember):
    def __init__(self, name, design_tool):
        super().__init__(name, "Designer")
        self.design_tool = design_tool

    def get_details(self):
        return f"{self.role}: {self.name}, Design Tool: {self.design_tool}"

# Demonstrate the usage
manager = Manager("John Doe", 5)
developer = Developer("Alice Smith", "Python")
designer = Designer("Bob Johnson", "Adobe Illustrator")

# Using the common method
team_members = [manager, developer, designer]
for member in team_members:
    print(member.get_details())
```

```python
#Que16
from abc import ABC, abstractmethod

class ElectronicDevice(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.powered_on = False

    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def shutdown(self):
        pass

class Television(ElectronicDevice):
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size

    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} TV is now powered on.")
            self.powered_on = True
        else:
            print(f"{self.brand} {self.model} TV is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} TV is now shutting down.")
            self.powered_on = False
        else:
            print(f"{self.brand} {self.model} TV is already powered off.")

class Laptop(ElectronicDevice):
    def __init__(self, brand, model, screen_size):
        super().__init__(brand, model)
        self.screen_size = screen_size

    def power_on(self):
        if not self.powered_on:
            print(f"{self.brand} {self.model} laptop is now powered on.")
            self.powered_on = True
        else:
            print(f"{self.brand} {self.model} laptop is already powered on.")

    def shutdown(self):
        if self.powered_on:
            print(f"{self.brand} {self.model} laptop is now shutting down.")
            self.powered_on = False
        else:
            print(f"{self.brand} {self.model} laptop is already powered off.")

# Demonstrate the usage
tv = Television("Sony", "Bravia", 55)
laptop = Laptop("Dell", "XPS", 13)

# Using common methods through abstraction
devices = [tv, laptop]
for device in devices:
    device.power_on()
    device.shutdown()
```

```python
#Que19
from abc import ABC, abstractmethod

class LibraryItem(ABC):
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    @abstractmethod
    def display_info(self):
        pass

    @abstractmethod
    def check_out(self):
        pass

    @abstractmethod
    def check_in(self):
        pass

class Book(LibraryItem):
    def __init__(self, title, author, genre):
        super().__init__(title, author)
        self.genre = genre

    def display_info(self):
        print(f"Book: {self.title} by {self.author}, Genre: {self.genre}")

    def check_out(self):
        if self.is_available:
            print(f"Checked out {self.title}.")
            self.is_available = False
        else:
            print(f"{self.title} is not available for checkout.")

    def check_in(self):
        if not self.is_available:
            print(f"Checked in {self.title}.")
            self.is_available = True
        else:
            print(f"{self.title} is already available.")

class DVD(LibraryItem):
    def __init__(self, title, director, duration):
        super().__init__(title, director)
        self.duration = duration

    def display_info(self):
        print(f"DVD: {self.title} directed by {self.author}, Duration: {self.duration} minutes")

    def check_out(self):
        if self.is_available:
            print(f"Checked out {self.title}.")
            self.is_available = False
        else:
            print(f"{self.title} is not available for checkout.")

    def check_in(self):
        if not self.is_available:
            print(f"Checked in {self.title}.")
            self.is_available = True
        else:
            print(f"{self.title} is already available.")

# Demonstrate the usage
book = Book("The Catcher in the Rye", "J.D. Salinger", "Fiction")
dvd = DVD("Inception", "Christopher Nolan", 148)

# Using common methods through abstraction
library_items = [book, dvd]
for item in library_items:
    item.display_info()
    item.check_out()
    item.check_in()
```

```python
#Que20
'''
Method abstraction in Python is a concept where a method is defined in a class without specifying its
implementation. The method is marked as abstract using the @abstractmethod decorator from the abc
module. This means that any concrete (non-abstract) subclass must provide an implementation for the 
abstract method. The primary goal of method abstraction is to define a common interface for a group 
of related classes, allowing them to share a common method signature without dictating the details of 
the implementation.
'''
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

# Usage
circle = Circle(5)
rectangle = Rectangle(4, 6)

print("Area of the circle:", circle.area())
print("Area of the rectangle:", rectangle.area())
```

```python
#Que6
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing: {self.title} by {self.artist} ({self.duration} seconds)")

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []

    def add_song(self, song):
        self.songs.append(song)

    def play(self):
        print(f"Playlist: {self.name}")
        for song in self.songs:
            song.play()

class MusicPlayer:
    def __init__(self):
        self.playlists = []

    def add_playlist(self, playlist):
        self.playlists.append(playlist)

    def play_all(self):
        print("Playing all playlists:")
        for playlist in self.playlists:
            playlist.play()

# Example Usage:
song1 = Song("Imagine", "John Lennon", 180)
song2 = Song("Bohemian Rhapsody", "Queen", 300)
song3 = Song("Hey Jude", "The Beatles", 240)

playlist1 = Playlist("Favorites")
playlist1.add_song(song1)
playlist1.add_song(song2)

playlist2 = Playlist("Rock Classics")
playlist2.add_song(song2)
playlist2.add_song(song3)

player = MusicPlayer()
player.add_playlist(playlist1)
player.add_playlist(playlist2)

player.play_all()
```

```python
#Que12
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def enroll(self, course):
        print(f"{self.name} enrolled in {course.course_name}")

class Instructor:
    def __init__(self, instructor_id, name):
        self.instructor_id = instructor_id
        self.name = name

    def assign_grade(self, student, course, grade):
        print(f"Instructor {self.name} assigned a grade of {grade} to {student.name} for {course.course_name}")

class CourseMaterial:
    def __init__(self, textbooks, syllabus):
        self.textbooks = textbooks
        self.syllabus = syllabus

class UniversityCourse:
    def __init__(self, course_code, course_name, instructor, students, course_material):
        self.course_code = course_code
        self.course_name = course_name
        self.instructor = instructor
        self.students = students
        self.course_material = course_material

    def display_course_info(self):
        print(f"Course: {self.course_code} - {self.course_name}")
        print(f"Instructor: {self.instructor.name}")
        print("Students Enrolled:")
        for student in self.students:
            print(f"- {student.name}")
        print("Course Material:")
        print(f"- Textbooks: {', '.join(self.course_material.textbooks)}")
        print(f"- Syllabus: {self.course_material.syllabus}")

# Example Usage:
student1 = Student(1, "Alice")
student2 = Student(2, "Bob")

instructor = Instructor(101, "Dr. Smith")

course_material = CourseMaterial(textbooks=["Math for Beginners", "Physics Fundamentals"],
                                 syllabus="Weekly schedule of topics and assignments")

students_enrolled = [student1, student2]

math_course = UniversityCourse(course_code="MATH101", course_name="Introduction to Mathematics",
                               instructor=instructor, students=students_enrolled, course_material=course_material)

# Display course information
math_course.display_course_info()

# Enroll students
for student in students_enrolled:
    student.enroll(math_course)

# Assign grades
instructor.assign_grade(student1, math_course, "A")
instructor.assign_grade(student2, math_course, "B")
```

```python
'''
Que15

Composition is a design principle in object-oriented programming that involves creating a class by combining or composing multiple simpler classes. In Python, composition is a powerful tool that enhances code maintainability and modularity in several ways:

Modularity:

Encapsulation of Components:

Composition allows you to encapsulate the implementation details of individual components within separate classes. Each component can be designed and implemented independently, promoting modularity.
Reusability:

Independent components created through composition are often reusable. They can be used in various contexts, leading to more modular and maintainable code. For example, a class representing a database connection could be reused across different parts of the system.
Flexibility and Decoupling:

Reduced Dependency on Implementation Details:

Composition reduces dependencies between classes, as objects interact through well-defined interfaces rather than relying on the details of each other's implementation. This decoupling improves flexibility, making it easier to modify or replace one component without affecting others.
Easy Component Swapping:

Since components are encapsulated within their own classes, swapping one component with another that adheres to the same interface becomes straightforward. This flexibility allows for the replacement of components without major modifications to the rest of the system.
Code Maintainability:

Isolation of Changes:

Changes to one component do not necessarily impact the entire system. With composition, you can isolate modifications to specific classes, reducing the risk of unintended side effects.
Easier Debugging and Testing:

Independent components are easier to debug and test. By focusing on the behavior of each component in isolation, developers can more easily identify and address issues. This isolation also facilitates unit testing.

Que17
Aggregation is a form of containment in object-oriented programming. 
It differs from composition in that it doesn't imply ownership. In composition, the parent entity
owns the child entity. In aggregation, the parent has a "has-a" relationship with the child entity.
'''
```

```python
'''
Que19
Factory Pattern:
Definition: The factory pattern provides an interface for creating objects in a super class but allows
subclasses to alter the type of objects that will be created.

Benefits: By using factories, you can dynamically change the type of objects created by a class,
providing flexibility in the object composition.
'''
class Shape:
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("Drawing a circle.")

class Square(Shape):
    def draw(self):
        print("Drawing a square.")

class ShapeFactory:
    def create_shape(self, shape_type):
        pass

class ConcreteShapeFactory(ShapeFactory):
    def create_shape(self, shape_type):
        if shape_type == "circle":
            return Circle()
        elif shape_type == "square":
            return Square()

# Dynamic object creation using factory
factory = ConcreteShapeFactory()
shape = factory.create_shape("circle")
shape.draw()
```

```python
#Que16
class Weapon:
    def __init__(self, name, damage):
        self.name = name
        self.damage = damage

    def attack(self):
        print(f"Attacking with {self.name}, dealing {self.damage} damage.")

class Armor:
    def __init__(self, name, defense):
        self.name = name
        self.defense = defense

    def absorb_damage(self, incoming_damage):
        absorbed_damage = min(incoming_damage, self.defense)
        self.defense -= absorbed_damage
        print(f"{self.name} absorbed {absorbed_damage} damage. Remaining defense: {self.defense}")

class Inventory:
    def __init__(self, items):
        self.items = items

    def add_item(self, item):
        self.items.append(item)
        print(f"Added {item} to the inventory.")

class GameCharacter:
    def __init__(self, name, health, weapon, armor, inventory):
        self.name = name
        self.health = health
        self.weapon = weapon
        self.armor = armor
        self.inventory = inventory

    def attack(self, target):
        target_health_before = target.health
        self.weapon.attack()
        target.receive_damage(self.weapon.damage)
        print(f"{target.name} health before: {target_health_before}, after: {target.health}")

    def equip_armor(self, armor):
        self.armor = armor
        print(f"{self.name} equipped {armor.name}.")

    def receive_damage(self, damage):
        if self.armor:
            self.armor.absorb_damage(damage)
        else:
            self.health -= damage
            print(f"{self.name} received {damage} damage. Remaining health: {self.health}")

# Example Usage:
sword = Weapon("Sword", 20)
shield = Armor("Shield", 15)

character1_inventory = Inventory(items=["Health Potion", "Mana Potion"])
character1 = GameCharacter("Knight", 100, sword, shield, character1_inventory)

axe = Weapon("Axe", 25)
armor = Armor("Plate Armor", 30)

character2_inventory = Inventory(items=["Scroll of Fireball", "Scroll of Invisibility"])
character2 = GameCharacter("Wizard", 80, axe, armor, character2_inventory)

# Battle between characters
character1.attack(character2)
character2.attack(character1)
character1.equip_armor(Armor("Chainmail", 20))
character2.attack(character1)

# Display characters' inventory
print(f"\n{character1.name}'s Inventory:")
for item in character1.inventory.items:
    print(f"- {item}")

print(f"\n{character2.name}'s Inventory:")
for item in character2.inventory.items:
    print(f"- {item}")
```

```python
#Que14
class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

class Dish:
    def __init__(self, name, description, ingredients):
        self.name = name
        self.description = description
        self.ingredients = ingredients

    def display_dish(self):
        print(f"{self.name}: {self.description}")
        print("Ingredients:")
        for ingredient in self.ingredients:
            print(f"- {ingredient.quantity} {ingredient.unit} of {ingredient.name}")

class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def display_menu(self):
        print(f"Menu: {self.name}")
        print("Dishes:")
        for dish in self.dishes:
            dish.display_dish()
            print()

class Restaurant:
    def __init__(self, name, menus):
        self.name = name
        self.menus = menus

    def display_restaurant(self):
        print(f"Welcome to {self.name} Restaurant")
        print("Available Menus:")
        for menu in self.menus:
            menu.display_menu()
            print()

# Example Usage:
ingredient1 = Ingredient("Chicken", 200, "grams")
ingredient2 = Ingredient("Rice", 150, "grams")
ingredient3 = Ingredient("Tomato", 2, "pieces")
ingredient4 = Ingredient("Onion", 1, "piece")

dish1 = Dish("Chicken Biryani", "A flavorful rice dish with chicken", [ingredient1, ingredient2, ingredient3, ingredient4])
dish2 = Dish("Vegetarian Pizza", "A classic pizza with assorted veggies", [ingredient2, ingredient3, ingredient4])

menu1 = Menu("Lunch Menu", [dish1, dish2])

ingredient5 = Ingredient("Salmon", 150, "grams")
ingredient6 = Ingredient("Asparagus", 100, "grams")
ingredient7 = Ingredient("Lemon", 1, "piece")

dish3 = Dish("Grilled Salmon", "Freshly grilled salmon with asparagus", [ingredient5, ingredient6, ingredient7])
dish4 = Dish("Caesar Salad", "Crispy salad with Caesar dressing", [ingredient3, ingredient4])

menu2 = Menu("Dinner Menu", [dish3, dish4])

restaurant = Restaurant("Fine Dining Restaurant", [menu1, menu2])

# Display restaurant information
restaurant.display_restaurant()
```

```python
#Que18
class Furniture:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Furniture: {self.name}")

class Appliance:
    def __init__(self, name):
        self.name = name

    def display_info(self):
        print(f"Appliance: {self.name}")

class Room:
    def __init__(self, name, furniture=None, appliances=None):
        self.name = name
        self.furniture = furniture or []
        self.appliances = appliances or []

    def add_furniture(self, furniture):
        self.furniture.append(furniture)

    def add_appliance(self, appliance):
        self.appliances.append(appliance)

    def display_info(self):
        print(f"\nRoom: {self.name}")
        print("Furniture:")
        for item in self.furniture:
            item.display_info()
        print("Appliances:")
        for item in self.appliances:
            item.display_info()

class House:
    def __init__(self, name, rooms):
        self.name = name
        self.rooms = rooms

    def add_room(self, room):
        self.rooms.append(room)

    def display_info(self):
        print(f"House: {self.name}")
        for room in self.rooms:
            room.display_info()

# Example Usage:
living_room = Room("Living Room", furniture=[Furniture("Sofa"), Furniture("Coffee Table")], appliances=[Appliance("TV")])
kitchen = Room("Kitchen", furniture=[Furniture("Dining Table"), Furniture("Chairs")], appliances=[Appliance("Refrigerator"), Appliance("Oven")])
bedroom = Room("Bedroom", furniture=[Furniture("Bed"), Furniture("Wardrobe")], appliances=[Appliance("Lamp")])

my_house = House("Cozy Home", rooms=[living_room, kitchen, bedroom])

# Display house information
my_house.display_info()
```

```python
#Que20
class Comment:
    def __init__(self, user, text):
        self.user = user
        self.text = text

    def display_info(self):
        print(f"Comment by {self.user}: {self.text}")

class Post:
    def __init__(self, user, content, comments=None):
        self.user = user
        self.content = content
        self.comments = comments or []

    def add_comment(self, comment):
        self.comments.append(comment)

    def display_info(self):
        print(f"\nPost by {self.user}: {self.content}")
        print("Comments:")
        for comment in self.comments:
            comment.display_info()

class User:
    def __init__(self, username, bio):
        self.username = username
        self.bio = bio
        self.posts = []

    def create_post(self, content):
        post = Post(user=self.username, content=content)
        self.posts.append(post)
        return post

    def display_info(self):
        print(f"\nUser: {self.username}")
        print(f"Bio: {self.bio}")
        print("Posts:")
        for post in self.posts:
            post.display_info()

class SocialMediaApp:
    def __init__(self, name, users=None):
        self.name = name
        self.users = users or []

    def add_user(self, user):
        self.users.append(user)

    def display_info(self):
        print(f"Social Media App: {self.name}")
        print("Users:")
        for user in self.users:
            user.display_info()

# Example Usage:
user1 = User(username="Alice123", bio="Travel enthusiast")
user2 = User(username="Bob456", bio="Foodie and chef")

post1 = user1.create_post("Exploring the mountains!")
comment1 = Comment(user="Charlie789", text="Looks amazing!")
post1.add_comment(comment1)

post2 = user2.create_post("Cooked a delicious meal today!")
comment2 = Comment(user="David101", text="Can you share the recipe?")
post2.add_comment(comment2)

app = SocialMediaApp(name="MySocialApp", users=[user1, user2])

# Display social media app information
app.display_info()
```
